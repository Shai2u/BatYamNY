```python
# Get a reference of the old script file and to street file
temp_folder='temp_files_trash'

#existing building
read_file = 'Existing_buildings.xlsx'
old_script = temp_folder + '/' +read_file
old_script_ds = pd.read_excel(old_script)

#street dictionary
strett_dic = 'street_id_dictionary.xlsx'
str_path = temp_folder + '/' +strett_dic
street_ds = pd.read_excel(str_path)

#extrapulated house size
extrapulated = 'apartment_sizes_extrapulated.xlsx'
ex_path = temp_folder + '/' +extrapulated
extrapulated_ds = pd.read_excel(ex_path)
```

```python
bld_plan = old_script_ds[['bld_Id','FuturePlanCode','purchase_p','rent_price']]
```
