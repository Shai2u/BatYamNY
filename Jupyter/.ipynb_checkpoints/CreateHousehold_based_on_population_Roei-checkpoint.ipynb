```python

def create_HH_pool_vetek(baalut,buil,percentage,num,vetek):
    #baalut = 1 (baal) or 0 (socher)
    #buil - is building data
    #percentage is:
    #columns=['old_poor_single','old_poor_married','old_rich_single','old_rich_married','young_poor_single','young_poor_married','young_rich_single','young_rich_married']

    #veetek is average vetek for baalut or socher (changes acculuativly)
    
    P=[]
    for ii in range(0,8):
      P.append(percentage.iloc[0,ii])
    
    HH_pool=pd.DataFrame()
    HH_pool['age']=[1,1,1,1,0,0,0,0] #1 old #0 young
    HH_pool['socioeconomic']=[1,1,0,0,1,1,0,0] #1 poor #2 rich
    HH_pool['baalut']=[baalut]*8       # 
    HH_pool['single']=[1,0,1,0,1,0,1,0]
    HH_pool['building']=[buil]*8
    HH_pool['vetek']=[vetek]*8
    
    #HH=pd.DataFrame()
    LLL=np.random.choice(8,num,p=P)

    HH=HH_pool.iloc[LLL]
    #for ii in range(0,len(LLL)):
    # g=HH_pool.iloc[ii,:]
    # HH=HH.append(g)
    #HH_pool=HH_pool.iloc[LLL,:]
    #HH_pool['percentage']=P
    #print((HH_pool))
    return HH

```

```python

# Functions
def analyze_pop(data):
    # All
    # old
    # single
    old_poor_single=len(data.loc[(data['age']==1) & (data['socioeconomic']==1) & (data['single']==1)])/len(data)
    old_poor_married=len(data.loc[(data['age']==1) & (data['socioeconomic']==1) & (data['single']==0)])/len(data)
    # poor
    old_rich_single=len(data.loc[(data['age']==1) & (data['socioeconomic']==0) & (data['single']==1)])/len(data)
    old_rich_married=len(data.loc[(data['age']==1) & (data['socioeconomic']==0) & (data['single']==0)])/len(data)
    # young
    young_poor_single=len(data.loc[(data['age']==0) & (data['socioeconomic']==1) & (data['single']==1)])/len(data)
    young_poor_married=len(data.loc[(data['age']==0) & (data['socioeconomic']==1) & (data['single']==0)])/len(data)
    # poor
    young_rich_single=len(data.loc[(data['age']==0) & (data['socioeconomic']==0) & (data['single']==1)])/len(data)
    young_rich_married=len(data.loc[(data['age']==0) & (data['socioeconomic']==0) & (data['single']==0)])/len(data)

    All=pd.DataFrame(columns=['old_poor_single','old_poor_married','old_rich_single','old_rich_married','young_poor_single','young_poor_married','young_rich_single','young_rich_married'])
    All.loc[0]=[old_poor_single]+[old_poor_married]+[old_rich_single]+[old_rich_married]+[young_poor_single]+[young_poor_married]+[young_rich_single]+[young_rich_married]

    # poor
    Poor = pd.DataFrame(columns=['poor','rich'])
    P = len(data.loc[data['socioeconomic']==1])/len(data)
    R = len(data.loc[data['socioeconomic']==0])/len(data)
    Poor.loc[0]=[P]+[R]

    # old/young
    Age = pd.DataFrame(columns=['young','old'])
    Y = len(data.loc[data['age']==0])/len(data)
    O = len(data.loc[data['age']==1])/len(data)
    Age.loc[0]=[P]+[R]

    # single
    Single= pd.DataFrame(columns=['single','married'])
    S = len(data.loc[data['single']==1])/len(data)
    M = len(data.loc[data['single']==0])/len(data)
    Single.loc[0]=[S]+[M]

    return (All,Poor,Age,Single)
```

```python

def create_synthetic_HH(HH_b,building_data,HH):
    #HH_b household form specific buidling
    #building_data - data of current building
    #HH population
    sochrim=(len(HH.loc[HH['baalut']==0]))/len(HH)  #percent of Socharim
    vetek_B=HH['vetek'].loc[HH['baalut']==1].mean() #vetek of people who own
    vetek_S=HH['vetek'].loc[HH['baalut']==0].mean() #vetek of people rent
    

    #Analyze Pops returns the properties of the population distribution see next line
    #columns=['old_poor_single','old_poor_married','old_rich_single','old_rich_married','young_poor_single','young_poor_married','young_rich_single','young_rich_married']
    [All_b,Poor,Age,Single] = analyze_pop(HH.loc[HH['baalut']==1])
    [All_s,Poor,Age,Single] = analyze_pop(HH.loc[HH['baalut']==0])
    
    S_HH=building_data['original_units']-len(HH_b) #if the number of Units is smaller than the number of Agents
    #Get the number of units needed
    if S_HH>0:
        #if Units need to be filled
        sochrim=round(S_HH*sochrim) #get the number of Socharim
        bealim=S_HH-sochrim #get thenumber of Bealim
        if bealim>0:
            #new_HH_B=create_HH_pool_vetek(1,building_data['bld_id'],All_b,int(new_bealim),vetek_B)
                new_HH_B=create_HH_pool_vetek(1,building_data['bld_id'],All_b,int(bealim),vetek_B)
            HH_b=HH_b.append(new_HH_B, ignore_index=True)
        if sochrim>0:
            #new_HH_B=create_HH_pool_vetek(0,building_data['bld_id'],All_s,int(new_bealim),vetek_S)
            new_HH_B=create_HH_pool_vetek(0,building_data['bld_id'],All_s,int(sochrim),vetek_S)
            HH_b=HH_b.append(new_HH_B, ignore_index=True)
    
    
    
    return HH_b


def create_synthetic_HH_0(HH,Existing_buildings):
    HH1=HH #save a copy of households
    
    for i in range(0,len(Existing_buildings)):
        #Iterate through existing buildings
        building_data=Existing_buildings.iloc[i] # Get data for specific building
        HH_b = HH1.loc[HH1['building']==building_data['bld_id']] # get agents from sepcific building
        
        HH_b=create_synthetic_HH(HH_b,building_data,HH) 
        HH1=HH1.append(HH_b, ignore_index=True)
    
        
    return(HH1)    
```

```python
Existing_buildings = pd.read_excel('Existing_buildings.xlsx')
HH = pd.read_excel('HH_temp.xlsx')
HH2=create_synthetic_HH_0(HH,Existing_buildings)
```
